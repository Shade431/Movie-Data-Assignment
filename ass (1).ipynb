```python
import pandas as pd

df = pd.read_csv('Movie-Data (1) (3).csv')
df.head()


```

```python
# 3. Show the shape, info, and description of the dataset
print("Shape of DataFrame:", df.shape)
print("\nGeneral Information:")
print(df.info())
print("\nDescription:")
print(df.describe(include='all'))
```

```python
# 4. Rename Columns

# Rename 'Runtime (Minutes)' to 'Runtime_Minutes' and 'Revenue (Millions)' to 'Revenue_Millions'
df.rename(columns={
    'Runtime (Minutes)': 'Runtime_Minutes',
    'Revenue (Millions)': 'Revenue_Millions'
}, inplace=True)
```

```python
# 5. Check for Missing Values
print("\nMissing values per column:")
print(df.isnull().sum())

# How many missing Descriptions?
print("\nNumber of missing Descriptions:", df['Description'].isnull().sum())
```

```python
# 6. Show the year with the maximum number of movies
most_movies_year = df['Year'].value_counts().idxmax()
print("\nYear with the maximum number of movies:", most_movies_year)
```

```python
# 7. Show the Top Ten Genre Categories
top_genres = df['Genre'].value_counts().head(10)
print("\nTop 10 Genres:")
print(top_genres)
```

```python
# 8. Show the Top Five and Least Five Voted Movies
sorted_votes = df.sort_values(by='Votes', ascending=False)
print("\nTop 5 Voted Movies:")
print(sorted_votes[['Title', 'Votes']].head(5))

print("\nLeast 5 Voted Movies:")
print(sorted_votes[['Title', 'Votes']].tail(5))
```

```python
# 9. Show the mean, standard deviation, skewness, and kurtosis of Revenues
print("\nRevenue Statistics:")
revenues = df['Revenue_Millions'].dropna()
print("Mean:", revenues.mean())
print("Standard Deviation:", revenues.std())
print("Skewness:", skew(revenues))
print("Kurtosis:", kurtosis(revenues))
```

```python
# 10. Show the 25%, 50%, 75%, and 90% quantiles of movie Ratings
print("\nQuantiles for Ratings:")
print(df['Rating'].quantile([0.25, 0.5, 0.75, 0.9]))
```

```python
# 11. Is there a relationship between Runtime and Rating?
runtime_rating_corr = df['Runtime_Minutes'].corr(df['Rating'])
print("\nCorrelation between Runtime and Rating:", runtime_rating_corr)
```

```python
# 12. Which parameter affects Revenue more: Votes or Rating?
revenue_votes_corr = df['Revenue_Millions'].corr(df['Votes'])
revenue_rating_corr = df['Revenue_Millions'].corr(df['Rating'])
print("\nCorrelation between Revenue and Votes:", revenue_votes_corr)
print("Correlation between Revenue and Rating:", revenue_rating_corr)
```

```python
# 13. Show all movies with Votes > 250,000
print("\nMovies with Votes greater than 250,000:")
print(df[df['Votes'] > 250000][['Title', 'Votes']])
```
